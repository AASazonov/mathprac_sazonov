```python
#   A quadratic equation axÂ² + bx + c = 0 has either 0, 1, or 2 distinct solutions for real values of x. Given a, b and c, you should return the number of solutions to the equation.
#   https://edabit.com/challenge/o2AKq4xy3nfZabKXL
#   2 points (Medium)

def solutions (a, b, c):
    D = b*b - 4*a*c
    if   (D > 0):
        return 2
    elif (D == 0):
        return 1
    else:
        return 0


Test.assert_equals(solutions(1, 0, -1), 2)
Test.assert_equals(solutions(1, 0, 0), 1)
Test.assert_equals(solutions(1, 0, 1), 0)
Test.assert_equals(solutions(200, 420, 800), 0)
Test.assert_equals(solutions(200, 420, -800), 2)
Test.assert_equals(solutions(1000, 1000, 0), 2)
Test.assert_equals(solutions(10000, 400, 4), 1)

#   Points: 2 of 15
```

```python
#   Create a function that uses bisection search to compute the approximative value of the square root of a number.
#   https://edabit.com/challenge/N5gBcQZQwJFwb8Z2k
#   4 points (Very Hard)

eps = 0.0001

def bisec_sqrt (n):
    return round(bisec_sqrt_rec(n, 0, n), 3)

def bisec_sqrt_rec (n, low, high):
    mid = (low + high)/2
    if   (n - mid*mid > eps):
        low = mid
        return bisec_sqrt_rec (n, low, high)
    elif ( mid*mid - n > eps):
        high = mid
        return bisec_sqrt_rec (n, low, high)
    else:
        return mid


Test.assert_equals(bisec_sqrt(69), 8.307)
Test.assert_equals(bisec_sqrt(16), 4.0)
Test.assert_equals(bisec_sqrt(12984771), 3603.439)
Test.assert_equals(bisec_sqrt(12.21), 3.494)
Test.assert_equals(bisec_sqrt(121.4), 11.018)

#   Points: 6 of 15
```

```python
#   Create a function that returns how many times it's been called previously. Do not use a global variable.
#   https://edabit.com/challenge/oo5zkiydtRB344ZcL
#   4 points (Very Hard)

def counter():
    counter.count += 1
    return counter.count

counter.count = -1



Test.assert_equals(counter(), 0, 'Example 1')
Test.assert_equals(counter(), 1, 'Example 2')
Test.assert_equals(counter(), 2, 'Example 3')

# Additional tests
for k in range(10):
	Test.assert_equals(counter(), k + 3)

#   Points: 10 of 15
```

```python
#   Create a function that takes the number of people n in a room and calculates the probability that any two people in the room share the same birthday. Round the result to two decimal places.
#   https://edabit.com/challenge/yssd27bpgZBAw3oJ6
#   5 points (Expert)
import math

def probability(n):
    p0 = 1.0
    k = 1
    for k in range(1, n):
        p0 *= (1 - k/365)
    return round(1-p0, 2)

Test.assert_equals(probability(7), 0.06)
Test.assert_equals(probability(15), 0.25)
Test.assert_equals(probability(12), 0.17)
Test.assert_equals(probability(14), 0.22)
Test.assert_equals(probability(16), 0.28)
Test.assert_equals(probability(28), 0.65)
Test.assert_equals(probability(35), 0.81)
Test.assert_equals(probability(31), 0.73)
Test.assert_equals(probability(11), 0.14)
Test.assert_equals(probability(19), 0.38)

#   Points: 15 of 15

#   Done!
```
